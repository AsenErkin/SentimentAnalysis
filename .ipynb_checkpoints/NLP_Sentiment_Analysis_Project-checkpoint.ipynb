## 3. Text Mining

Text is the most unstructured form of all the available data, therefore various types of noise are present in it. This means that the data is not readily analyzable without any pre-processing. The entire process of cleaning and standardization of text, making it noise-free and ready for analysis is known as **text preprocessing**.

The three key steps of text preprocessing:

- **Tokenization:**
This step is one of the top priorities when it comes to working on text mining. Tokenization is essentially splitting a phrase, sentence, paragraph, or an entire text document into smaller units, such as individual words or terms. Each of these smaller units are called tokens.

- **Noise Removal:**
Any piece of text which is not relevant to the context of the data and the end-output can be specified as the noise.
For example  language stopwords (commonly used words of a language  is, am, the, of, in etc), URLs or links, upper and lower case differentiation, punctuations and industry specific words. This step deals with removal of all types of noisy entities present in the text.


- **Lexicon Normalization:**
Another type of textual noise is about the multiple representations exhibited by single word.
For example  play, player, played, plays and playing are the different variations of the word  play. Though they mean different things, contextually they all are similar. This step converts all the disparities of a word into their normalized form (also known as lemma). 
There are two methods of lexicon normalisation; **[Stemming or Lemmatization](https://www.guru99.com/stemming-lemmatization-python-nltk.html)**. Lemmatization is recommended for this case, because Lemmatization as this will return the root form of each word (rather than just stripping suffixes, which is stemming).

As the first step change text to tokens and convertion all of the words to lower case.  Next remove punctuation, bad characters, numbers and stop words. The second step is aimed to normalization them throught the Lemmatization method. 


***Note:*** *Use the functions of the ***[nltk Library](https://www.guru99.com/nltk-tutorial.html)*** for all the above operations.*



### Tokenization, Noise Removal, Lexicon Normalization

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
        stop_words.remove(i)
```

```python
def cleaning(data):
    
    
    #1. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")
    
    #2. Tokenize
    text_tokens = word_tokenize(data.lower())
    
    #3. Remove numbers
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #4. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #5. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
     
    #6. joining
    return " ".join( text_cleaned)

```

```python
cleaning_text = df1["review"].apply(cleaning)
cleaning_text.head()
```

```python
# check them
cleaning_text[cleaning_text.str.contains("@")]
```

Machine learning algorithms most often take numeric feature vectors as input. Thus, when working with text documents, you need a way to convert each document into a numeric vector. This process is known as text vectorization. Commonly used vectorization approach that you will use here is to represent each text as a vector of word counts.

At this moment, you have your review text column as a token (which has no punctuations and stopwords). You can use Scikit-learns CountVectorizer to convert the text collection into a matrix of token counts. You can imagine this resulting matrix as a 2-D matrix, where each row is a unique word, and each column is a review.

### Train - Test Split

To run machine learning algorithms we need to convert text files into numerical feature vectors. We will use bag of words model for our analysis.

First we spliting the data into train and test sets:

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split

X = df2["review"]
y = df2["recommendation"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=101)
```

In the next step we create a numerical feature vector for each document:

### Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(preprocessor=cleaning, min_df=3, max_df=0.9)
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf_vectorizer = TfidfVectorizer(preprocessor=cleaning, min_df=3, max_df=0.9)
X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)
```

### Eval Function

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

### Fixing token counts of all documents

```python
# find the longest token
# For every sentence...
number_tokens_sentences = []
for sent in X:

    # Tokenize the text and add `[CLS]` and `[SEP]` tokens.
    
    input_ids = tokenizer.encode(sent, add_special_tokens=True)
    number_tokens_sentences.append(len(input_ids))

print('Max sentence length: ', max(number_tokens_sentences))
# model takes max 512 tokens; since the max is 162, we can continue with this to avoid information loss
```

### Transformation Vectors to Matrices

```python
# with the asa.ki function we will transform each setup into the first 3 embeddings and then pull the input and attention mask with a for loop
def transformation(X):
  # set array dimensions
  seq_len = 162

  # We create the Xids empty list for input_ids and the Xmask empty list for attention_masks.
  Xids = []
  Xmask = []

    
  for sentence in X: # We extract all comments from the corpus in order.

      tokens = tokenizer.encode_plus(sentence, max_length=seq_len, truncation=True,
                                      padding='max_length', add_special_tokens=True) 
      # we convert each comment (sentence) we've extracted into vectors.
      
      # assign tokenized outputs to respective rows in numpy arrays
      Xids.append(tokens['input_ids']) # append the input_ids vector corresponding to each comment to the Xids list.
      Xmask.append(tokens['attention_mask']) # append the attention_mask vector Xmask corresponding to each comment.

  return np.array(Xids), np.array(Xmask) # We convert the Xids and Xmask matrices (number of comments, 162) to np.array format for faster execution.
```

```python
Xids_train, Xmask_train = transformation(X_train)

# We organize our 2 separate matrices for our train set.

Xids_test, Xmask_test = transformation(X_test)
# For our test set, we also organize our 2 separate matrices.

# our inputs are no longer specified by a single variable, but by 2 separate matrices. 
```

```python
print("Xids_train.shape  :", Xids_train.shape)
print("Xmask_train.shape :", Xmask_train.shape)
print("Xids_test.shape   :", Xids_test.shape)
print("Xmask_test.shape  :", Xmask_test.shape)
```

```python
labels_train = y_train.reshape(-1,1)
labels_train

# BERT models also require the target as a matrix. Therefore, we convert our y_train target into a matrix of size (len(y_train), 1).
```

### Transformation Matrix to Tensorflow tensor

```python
import tensorflow as tf

dataset_train = tf.data.Dataset.from_tensor_slices((Xids_train, Xmask_train, labels_train))
dataset_train

# The last step before feeding the data to the BERT models is to convert the matrices we obtained above into grouped Tensorflow tensors.
```

```python
dataset_test = tf.data.Dataset.from_tensor_slices((Xids_test, Xmask_test, labels_test))
dataset_test

# As we did with the train data, we will convert our 3 separate matrices into grouped Tensorflow tensors for the test data.
```

```python
def map_func(Tensor_Xids, Tensor_Xmask, Tensor_labels):
    # we convert our three-item tuple into a two-item tuple where the input item is a dictionary
    return {'input_ids': Tensor_Xids, 'attention_mask': Tensor_Xmask}, Tensor_labels
  
# It cannot know which of the tensors we obtained above is 'input_ids', which is 'attention_mask' and which is labels model. For this reason
# we need to provide this information to the model. For this we define the function above.
```

```python
dataset_train = dataset_train.map(map_func)
dataset_test = dataset_test.map(map_func) 
```

### Creating optimization

```python
!pip install -q tf-models-official
```

```python
from official.nlp import optimization  
epochs = 3 
#batch_size = 32
steps_per_epoch = len(train_ds) # steps_per_epoch: how many 32 packets will be fed to the model in each epoch.

num_train_steps = steps_per_epoch * epochs
num_warmup_steps = int(0.1*num_train_steps)

init_lr = 2e-5  # 3e-5, 5e-5  # best practice is one of these 3. since our data is small, let's start with the smallest steps
optimizer= optimization.create_optimizer(init_lr=init_lr,
                                          num_train_steps=num_train_steps,
                                          num_warmup_steps=num_warmup_steps,
                                          optimizer_type='adamw')  # nlp has its own optimizers that give better results, therefore 
                                          # above we downloaded the nlp official and imported the optimizer from here and from these we imported a different one called adamw
                                          # we use an optimizer
```

### Creating Model with TPU

```python
def create_model():
    from transformers import TFBertModel
    from tensorflow.keras.layers import Input, Dropout, Dense, BatchNormalization
    from tensorflow.keras import Model

    model = TFBertModel.from_pretrained("bert-base-uncased")

    input_ids = Input(shape=(162,), name='input_ids', dtype='int32')
    attention_mask = Input(shape=(162,), name='attention_mask', dtype='int32')

    embeddings = model.bert(input_ids=input_ids, attention_mask=attention_mask)["pooler_output"] 
    # pooler output, pre-trained modelin tokenleri hafizasinda tuttugu layer. burdaki embeddingsleri asagidaki dene layera aktaracagiz

    x = Dense(162, activation='relu')(embeddings)
    x = BatchNormalization()(x)
    x = Dropout(0.1)(x) #0.1 oran developerlar tarafndan tavsiye edilen orandr. 
    y = Dense(1, activation='sigmoid', name='outputs')(x) 


    return Model(inputs=[input_ids, attention_mask], outputs=y)
    # son aamada elde ettiimiz input ve outputlar Model fonksiyonu iine tanmlanr.
```

```python
# All the operations we want to do by taking advantage of the speed of the 8 TPU cores assigned to the strategy variable can be done with the strategy.scope() code block. 
# We define it.

with strategy.scope():

  from tensorflow.keras.losses import BinaryCrossentropy
  from tensorflow.keras.metrics import Recall

  loss = BinaryCrossentropy()
  recall = Recall()
  model3 = create_model()  

model3.compile(optimizer=optimizer, loss=loss, metrics=[recall])
```

```python
# Let's test our model by taking some prediction and end this project

# We apply the same process of transforming train and test data into grouped tensorflow tensors to the observations (obs) that we will predict respectively.

def prep_data(text):
  import tensorflow as tf
  
  # 1. We first transform the new observations we will predict into token embeddings (input_ids) and attention_mask matrices.
  # Note that you need to run the transformation function before you can use this function.
  Xids_obs, Xmask_obs = transformation(text) 
  
  # 2. Xids_obs, Xmask_obs matrixlerimizi gruplandrlm tenserflow tensrlerine dntryoruz.
  dataset_obs = tf.data.Dataset.from_tensor_slices((Xids_obs, Xmask_obs))
  
  # We name the transformed grouped tensorflow tensors as "input_ids" and "attention_mask" for the model to recognize them.
  def map_func(Tensor_Xids, Tensor_Xmask):
       return {'input_ids': Tensor_Xids, 'attention_mask': Tensor_Xmask}
  
  dataset_obs = dataset_obs.map(map_func)

  # 4. As the last step, we are grouping our grouped tensorflow tensors into packets of 32 as in the train data. Otherwise we will get a shape mismatch error. 
  batch_size = 32 # batch_size applied to the training data.
  obs_ds = dataset_obs.batch(batch_size)

  return obs_ds 
```
